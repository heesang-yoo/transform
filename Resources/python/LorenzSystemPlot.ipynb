```python
from buildingspy.io.outputfile import Reader
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
# Specify location of output file and load the file
matFile = '../../../../../Dymola/LorenzSystem_Test.mat'
r = Reader(matFile,'dymola')

# Save file(true/false) and location to save
togSave = False
figPath = '../Images/LorenzSystemFig.png'
```

```python
# Retrieve result data (i.e., time dependent)
(time, xs) = r.values('lorenzSystem.x')
ys = r.values('lorenzSystem.y')[1]
zs = r.values('lorenzSystem.z')[1]
```

```python
# Generatre figure
fig = plt.figure()
ax = fig.gca(projection='3d')

ax.plot(xs, ys, zs, lw=0.5)
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("Lorenz System")

plt.show()
```

```python
if togSave == True:
    fig.savefig(figPath,dpi=400)
```
